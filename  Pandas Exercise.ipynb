# Join & Merge

<img src="joins.jpg" width=900 >


```python
# joins are the basic tool for querying databases
# relational databases are in fact based in joins
# with Data Frames we can do exactly the same
# let's first have two hypothetical tables

df1 = pd.DataFrame({'name':['John', 'George','Ringo'],'color': ['Blue','Blue','Purple']})
df2 = pd.DataFrame({'name':['Paul','George','Ringo'],'carcolor':['Red','Blue',np.nan]},index=[3,1,2])
df1
df2
```

```python
# our first operation is not a join, 
# but a concatenation of the two tables
# by defalt concat preserves index values
pd.concat([df1,df2])

# but we can ignore them if we want
pd.concat([df1,df2], ignore_index=True)

# we can also concat by rows instead of columns
pd.concat([df1,df2],axis=1)
```

```python
d1=pd.DataFrame({'name':['A','B','B','C'],'size':[1,2,3,4]})

d2=pd.DataFrame({'name':['C','C','D','D'],'value':[10,9,8,7]},index=[3,1,2,4])

d1
d2

d1.merge(d2, on='name')                    # inner join
d1.merge(d2, how='outer', on='name')       #outer join
d1.merge(d2, how='left',  on='name')       #left join
d1.merge(d2, how='right', on='name')       # right join
```

```python
#voluntary
#read table
times=pd.read_csv("timesData.csv")
shanghai=pd.read_csv("shanghaiData.csv")

#times.head()

#drop NaNs in ratings 
times["total_score"]=pd.to_numeric(times['total_score'],errors="coerce")
times.dropna(subset=["total_score"], inplace=True)

#take only 2016 to avoid duplicate rating
ti=times[times["year"]==2016]

#determine top 10 countries by average total ratings
top_10=ti[["country","total_score"]].groupby(by="country").mean().sort_values(by="total_score",ascending=False).head(10)
top_50=top_10.merge(ti,on="country").sort_values(by="total_score_y",ascending=False).head(50)
top_50.drop("world_rank",axis=1,inplace=True)
#top_50.shape

#top_50

sns.scatterplot(x="total_score_y",y="country",data=top_50)

    
```

```python
sns.boxplot(x="total_score_y",y="country",data=top_50)
```

```python
#Are their rankings clustered? Are there very few good ones or there is a high dispersion.
#there is a high dispersion among rankings, as the countries with the highest university rankings also have ratings with lower scores. 
#the countries with low dispersion usually score lower on the total score as well. 
```
